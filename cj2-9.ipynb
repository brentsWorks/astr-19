Create a Jupyter notebook, import matplotlib. 

```python
import matplotlib.pyplot as plt
import numpy as np
```

Use numpy to pull 1000 random numbers distributed uniformly between [0,1].

```python
x = np.random.uniform(0, 1, 1000)
```

Histogram the random numbers into 100 bins, and plot the histogram. Label your axes and save the figure as a PDF. 

```python
f, ax = plt.subplots(1,1,figsize=(8,8))

ax.hist(x,facecolor='red',bins=100,edgecolor='black',alpha=0.5)
ax.set_xlabel('Values',fontsize=14)
ax.set_ylabel('Frequency',fontsize=14)

plt.savefig('histogram_plot.png',bbox_inches='tight',dpi=400)
```
