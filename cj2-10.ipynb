Create a Jupyter notebook, import matplotlib.

```python
import matplotlib.pyplot as plt
import numpy as np
```

Select logarithmic distribution and pull 1000 random numbers from that distribution. 


```python
p = 0.1
size = 1000
x = np.random.geometric(p, size)
```

Histogram the random numbers into 100 bins, and plot the histogram. Label your axes and save the figure as a PDF. 


```python
f, ax = plt.subplots(1,1,figsize=(8,8))

ax.hist(x,facecolor='red',bins=100,edgecolor='black',alpha=0.5)
ax.set_xlabel(f'# Of Trials Until First Success with p = {p}',fontsize=12)
ax.set_ylabel('Frequency',fontsize=12)
plt.title('Geometric Distribution Histogram')

plt.savefig('geometric_histogram_plot.png',bbox_inches='tight',dpi=400)
```
